```python
def dist_map(som, distSquared = False):
    
    w = som.get_weights()
    
    map_size = np.array([som.get_weights().shape[0],som.get_weights().shape[1]])
    
    
    neur=np.zeros(map_size)
    
    mp,mp_aux=grid(tam=map_size,g='hexagonal')

    cmap_aux = plt.cm.YlOrRd
   # cmap_aux = cmap_aux.reversed()
    
    l2=0
    for l in range(np.shape(mp)[0]-1):
        for c in range(np.shape(mp)[1]):
            plt.plot([mp[l,c,1],mp_aux[l2,c,1]],[mp[l,c,0],mp_aux[l2,c,0]],color='white')
            plt.plot([mp[l,c,1],mp_aux[l2+1,c,1]],[mp[l,c,0],mp_aux[l2+1,c,0]],color='white')
            plt.plot([mp[l+1,c,1],mp_aux[l2,c,1]],[mp[l+1,c,0],mp_aux[l2,c,0]],color='white')
            plt.plot([mp[l+1,c,1],mp_aux[l2+1,c,1]],[mp[l+1,c,0],mp_aux[l2+1,c,0]],color='white')
        l2=l2+2
    l2=0
    for l in range(0,np.shape(mp)[0]-1,2):
        for c in range(1,np.shape(mp)[1]):
            plt.plot([mp[l,c,1],mp_aux[l2,c-1,1]],[mp[l,c,0],mp_aux[l2,c-1,0]],color='white')
            plt.plot([mp[l+1,c-1,1],mp_aux[l2+1,c,1]],[mp[l+1,c-1,0],mp_aux[l2+1,c,0]],color='white')
        l2=l2+4
    l2=2
    for l in range(1,np.shape(mp)[0]-1,2):
        for c in range(0,np.shape(mp)[1]-1):
            plt.plot([mp[l,c,1],mp_aux[l2,c+1,1]],[mp[l,c,0],mp_aux[l2,c+1,0]],color='white')
            plt.plot([mp[l+1,c+1,1],mp_aux[l2+1,c,1]],[mp[l+1,c+1,0],mp_aux[l2+1,c,0]],color='white')
        l2=l2+4


    #p2=np.cross([mp[1,0,1],mp[1,0,0],1],[mp_aux[1,0,1],mp_aux[1,0,0],1])
    #xx=np.array([mp[1,0,1],mp_aux[1,0,1]])
    #plt.plot(xx,-(xx*p2[0]+p2[2])/p2[1],color='black')

    maxi=0
    if distSquared:
        mini=np.square(distance.euclidean(w[0,0,:],w[1,0,:]))
    else:
        mini=distance.euclidean(w[0,0,:],w[1,0,:])
    
    for l in range(0,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]):
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c,:]))
        
            
            if d>maxi:
                maxi=d
            if d<mini:
                mini=d
    
    for l in range(1,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]-1):
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c+1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c+1,:]))
            
            if d>maxi:
                maxi=d
            if d<mini:
                mini=d
                
    for l in range(0,np.shape(mp)[0]-1,2):
        for c in range(1,np.shape(mp)[1]):
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c-1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c-1,:]))
            
            
            if d>maxi:
                maxi=d
            if d<mini:
                mini=d
            
    for l in range(1,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]):
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c,:]))
            
            if d>maxi:
                maxi=d
            if d<mini:
                mini=d
            
    for l in range(1,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]-1):
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l,c+1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l,c+1,:]))
            
            
            if d>maxi:
                maxi=d
            if d<mini:
                mini=d
            
    for l in range(2,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]-1):
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l,c+1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l,c+1,:]))
            
            if d>maxi:
                maxi=d
            if d<mini:
                mini=d
            
    l2=0
    
    for l in range(0,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]):
            
            
            p2=np.cross([mp[l,c,1],mp[l,c,0],1],[mp_aux[l2,c,1],mp_aux[l2,c,0],1])
            xx=np.array([mp[l,c,1],mp_aux[l2+1,c,1],mp[l+1,c,1]])
            yy=-(xx*p2[0]+p2[2])/p2[1]
            
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c,:]))
            
            
            z=np.array([maxi,mini,d])
            normal = plt.Normalize(z.min(), z.max())
            cmap = cmap_aux(normal(z))
            
            plt.fill_between(xx,[mp[l,c,0],mp_aux[l2+1,c,0],mp[l+1,c,0]],yy,color=cmap[2][:])
            plt.plot([mp[l,c,1],mp[l+1,c,1]],[mp[l,c,0],mp[l+1,c,0]],color='black',lineWidth=0.5,lineStyle='-.')
        l2=l2+4

    l2=2
    for l in range(1,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]-1):

            p2=np.cross([mp[l,c,1],mp[l,c,0],1],[mp_aux[l2,c+1,1],mp_aux[l2,c+1,0],1])
            xx=np.array([mp[l,c,1],mp_aux[l2+1,c,1],mp[l+1,c+1,1]])
            yy=-(xx*p2[0]+p2[2])/p2[1]

            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c+1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c+1,:]))
        
            z=np.array([maxi,mini,d])
            normal = plt.Normalize(z.min(), z.max())
            cmap = cmap_aux(normal(z))
            
            plt.fill_between(xx,[mp[l,c,0],mp_aux[l2+1,c,0],mp[l+1,c+1,0]],yy,color=cmap[2][:])
            plt.plot([mp[l,c,1],mp[l+1,c+1,1]],[mp[l,c,0],mp[l+1,c+1,0]],color='black',lineWidth=0.5,lineStyle='-.')
        l2=l2+4

    l2=0
    for l in range(0,np.shape(mp)[0]-1,2):
        for c in range(1,np.shape(mp)[1]):

            p2=np.cross([mp[l,c,1],mp[l,c,0],1],[mp_aux[l2,c-1,1],mp_aux[l2,c-1,0],1])
            xx=np.array([mp[l,c,1],mp_aux[l2+1,c,1],mp[l+1,c-1,1]])
            yy=-(xx*p2[0]+p2[2])/p2[1]

            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c-1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c-1,:]))
            
            z=np.array([maxi,mini,d])
            normal = plt.Normalize(z.min(), z.max())
            cmap = cmap_aux(normal(z))
            
            plt.fill_between(xx,[mp[l,c,0],mp_aux[l2+1,c,0],mp[l+1,c-1,0]],yy,color=cmap[2][:])
            plt.plot([mp[l,c,1],mp[l+1,c-1,1]],[mp[l,c,0],mp[l+1,c-1,0]],color='black',lineWidth=0.5,lineStyle='-.')
        l2=l2+4

    l2=2
    for l in range(1,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]):

            p2=np.cross([mp[l,c,1],mp[l,c,0],1],[mp_aux[l2,c,1],mp_aux[l2,c,0],1])
            xx=np.array([mp[l,c,1],mp_aux[l2+1,c,1],mp[l+1,c,1]])
            yy=-(xx*p2[0]+p2[2])/p2[1]
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l+1,c,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l+1,c,:]))
            
            z=np.array([maxi,mini,d])
            normal = plt.Normalize(z.min(), z.max())
            cmap = cmap_aux(normal(z))

            plt.fill_between(xx,[mp[l,c,0],mp_aux[l2+1,c,0],mp[l+1,c,0]],yy,color=cmap[2][:])
            plt.plot([mp[l,c,1],mp[l+1,c,1]],[mp[l,c,0],mp[l+1,c,0]],color='black',lineWidth=0.5,lineStyle='-.')
        l2=l2+4

    l2=2
    for l in range(1,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]-1):
            
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l,c+1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l,c+1,:]))
            
            z=np.array([maxi,mini,d])
            normal = plt.Normalize(z.min(), z.max())
            cmap = cmap_aux(normal(z))
            
            plt.fill_between([mp[l,c,1],mp_aux[l2,c+1,1],mp[l,c+1,1]],[mp[l,c,0],mp_aux[l2,c+1,0],mp[l,c+1,0]],[mp[l,c,0],mp_aux[l2-1,c+1,0],mp[l,c,0]],color=cmap[2][:])
            plt.plot([mp[l,c,1],mp[l,c+1,1]],[mp[l,c,0],mp[l,c+1,0]],color='black',lineWidth=0.5,lineStyle='-.')
        l2=l2+4

    l2=4
    for l in range(2,np.shape(mp)[0]-1,2):
        for c in range(np.shape(mp)[1]-1):
            
            
            if distSquared:
                d=np.square( distance.euclidean(w[l,c,:],w[l,c+1,:]))
            else:
                d=(distance.euclidean(w[l,c,:],w[l,c+1,:]))
            

            z=np.array([maxi,mini,d])
            normal = plt.Normalize(z.min(), z.max())
            cmap = cmap_aux(normal(z))
            
            plt.fill_between([mp[l,c,1],mp_aux[l2,c,1],mp[l,c+1,1]],[mp[l,c,0],mp_aux[l2,c,0],mp[l,c+1,0]],[mp[l,c,0],mp_aux[l2-1,c,0],mp[l,c,0]],color=cmap[2][:])
            plt.plot([mp[l,c,1],mp[l,c+1,1]],[mp[l,c,0],mp[l,c+1,0]],color='black',lineWidth=0.5,lineStyle='-.')
        l2=l2+4

    plt.plot(mp[:,:,1],mp[:,:,0],'h',markerfacecolor='black',markersize=10,markeredgecolor='white')  
    plt.gca().set_aspect('equal')
   
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)


    cbar = plt.colorbar(cm.ScalarMappable(norm=normal, cmap=cmap_aux), ticks=[mini,maxi], cax=cax)

    cbar.set_ticklabels(['Min Dist', 'Max Dist'])

    plt.draw()
```

```python
def grid(tam,g='hexagonal'):
    if g=='hexagonal':
        mp=np.zeros((tam[0],tam[1],2))
        mp_aux=np.zeros((2*(tam[0]-1),tam[1],2))
       # mp_aux=np.zeros((4,5))
        auxL=0
        indLaux=0
        l=0
        for lAux in range(tam[0]):
            aux=lAux/2
            auxC=aux-int(aux)
            if auxC==0:
                auxC_aux_1=0.5
                auxC_aux_2=0
            else:
                auxC_aux_1=0
                auxC_aux_2=0.5
            for c in range(tam[1]):
                mp[lAux,c,0]=auxL
                mp[lAux,c,1]=auxC
                if indLaux+1<2*(tam[0]-1) and c<tam[1]:
                    mp_aux[indLaux,c,0]=auxL+(np.cos(np.pi/6))/3
                    mp_aux[indLaux,c,1]=auxC_aux_1
                    mp_aux[indLaux+1,c,0]=auxL+2*(np.cos(np.pi/6))/3
                    mp_aux[indLaux+1,c,1]=auxC_aux_2
                    auxC_aux_1=auxC_aux_1+1
                    auxC_aux_2=auxC_aux_2+1
                auxC=auxC+1
            indLaux=indLaux+2
            auxL=auxL+1
            l+=np.cos(np.pi/6)
    return mp, mp_aux
```

```python
def neurons_classes(som, data, labels):


    map_size = np.array([som.get_weights().shape[0],som.get_weights().shape[1]])
    mp, mp_aux=grid(map_size)
    r = 0.5
    for l in range(np.shape(mp)[0]):
        for c in range(np.shape(mp)[1]):

            y = mp[l,c,0]
            x = mp[l,c,1]

            for i in range(6):

                y1 = r * np.cos(i*np.pi/3) + y
                x1 = r * np.sin(i*np.pi/3) + x

                y2 = r * np.cos((i+1)*np.pi/3) + y
                x2 = r * np.sin((i+1)*np.pi/3) + x

                plt.plot([x1,x2],[y1,y2],color='black')

    #fig = plt.gcf()


    cmap = plt.cm.get_cmap('gist_rainbow', len(np.unique(labels)))


    w=np.zeros((map_size[0],map_size[1],10))

    for cnt, xx in enumerate(data):
        w_aux = som.winner(xx)
        w[w_aux[0], w_aux[1], labels[cnt] ] += 1  # getting the winner
    
    purity = 0
    entropy = 0
    c = np.argmax(w,axis=2)
    for i in range(map_size[0]):
        for j in range(map_size[1]):        
            if not (w[j,i,c[j,i]]==0):
                plt.annotate((c[j,i]).astype(int), [mp[j,i,1]-0.2,mp[j,i,0]-0.2], weight='bold', color = cmap(c[j,i]), size=10)
                purity += w[j,i,c[j,i]]
                entropy += - np.sum( [(w[j,i,k] * np.log2( w[j,i,k] / np.sum(w[j,i,:]) )) for k in np.arange(w.shape[2]) if not(w[j,i,k]==0) ] ) 
    plt.axis('off')
    plt.gca().set_aspect('equal')
    
    purity /= len(data)
    entropy /= len(data)
    
    return purity, entropy
```
