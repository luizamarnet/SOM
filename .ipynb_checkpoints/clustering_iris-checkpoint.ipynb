```python
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt

import import_ipynb
from utils import *
from som_plots import *
```

```python
data_arq = 'D:\Luiza\OneDrive\Documentos\Doutorado\MeusProjetos\SOM\\archive\Iris.csv'
data, labels = read_and_preprocess_dataset (data_arq )
```

```python
som = MiniSom(10,8, 4, sigma=1.5, learning_rate=.01, activation_distance='euclidean', topology='hexagonal',neighborhood_function='gaussian', random_seed=None)
print("Training...")
som.train_batch(data, 100000, verbose=True)  # random training
print("\n...ready!")
```

```python
# use different colors and markers for each label
markers = ['o', 's', 'D']
colors = ['C0', 'C1', 'C2']
for cnt, xx in enumerate(data):
    w = som.winner(xx)  # getting the winner
    # palce a marker on the winning position for the sample xx
    plt.plot(w[1]+.5,w[0]+.5, markers[labels[cnt]], markerfacecolor='None',
             markeredgecolor=colors[labels[cnt]], markersize=12, markeredgewidth=2)
#plt.axis([0, 7, 0, 7])
#plt.savefig('resulting_images/som_iris.png')
plt.gca().set_aspect('equal')
plt.show()
```
